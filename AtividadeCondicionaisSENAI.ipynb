Exercício 12 - Crie um programa que peça ao usuário dois números e uma operação (adição, subtração, multiplicação, divisão, exponenciação ou radiciação). Execute a operação escolhida e imprima o resultado. (não utilizar bibliotecas, utilizar apenas os operadores aritméticos).

```python
class Calculadora:
  def __init__(self,num1,num2):
    self.num1 = num1
    self.num2 = num2
  def resultado(self,ope):
    if ope == 1:
      return self.num1 + self.num2
    if ope == 2:
      return self.num1 - self.num2
    if ope == 3:
      return self.num1 * self.num2
    if ope == 4:
      return self.num1 / self.num2
    return "Opção invalida"

def menu():
  return (
    "1 - Somar +\n"
    "2 - Subtrair -\n"
    "3 - Multiplica x\n"
    "4 - Dividir /\n"
  )
if __name__ == "__main__":
  print(menu())
  try:
    ope = int(input("Informe a operação: "))
    num1 = float(input("Informe um numero: "))
    num2 = float(input("Informe um segundo numero: "))

    calcular = Calculadora(num1=num1,num2=num2)
    result = calcular.resultado(ope=ope)
    print(result)
  except ZeroDivisionError:
    print("Erro divisao por zero")
```

Exercício 20 - Peça ao usuário seu peso (em kg) e altura (em metros). Calcule o IMC usando a fórmula IMC = peso / (altura * altura) e classifique o resultado como "Abaixo do peso", "Peso normal", "Sobrepeso" ou "Obesidade", de acordo com a tabela de classificação do IMC.


```python
class Imc:
  def __init__(self,peso,altura):
    self.peso = peso
    self.altura = altura

  def verificar_imc(self):
    imc = self.peso / self.altura ** 2

    if imc <= 18.5:
      return f"Seu IMC: {imc:.2f} (Magreza)"
    if imc >= 18.5 and imc <= 24.9:
      return f"Seu IMC: {imc:.2f} (Peso normal)"
    if imc >= 25.0 and imc <= 29.9:
      return f"Seu IMC: {imc:.2f} (Sobre Peso)"
    if imc >= 30.0 and imc <= 39.9:
      return f"Seu IMC: {imc:.2f} (Obesidade)"
    return f"Seu IMC: {imc:.2f} (Obesidade Grave)"


if __name__ == "__main__":
  peso = float(input("Informe seu peso: "))
  altura = float(input("Informe sua altura: "))

  calcular = Imc(peso=peso,altura=altura)
  resultado = calcular.verificar_imc()
  print(resultado)
```
